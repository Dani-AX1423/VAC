```python
import spacy
nlp = spacy.load("en_core_web_sm")
```

```python
print(nlp.Defaults.stop_words)
```

```python
nlp.vocab['mystery'].is_stop
```

```python
nlp.Defaults.stop_words.add('btw')
nlp.vocab['btw'].is_stop = True
```

```python
from spacy.matcher import Matcher
matcher = Matcher(nlp.vocab)
```

```python
p1 = [{'LOWER':'solarpower'}]
p2 = [{'LOWER':'solar'},{'LOWER':'power'}]
p3 = [{'LOWER':'solar'},{'IS_PUNCT':True},{'LOWER':'power'}]
```

```python
matcher.add('SolarPower',[p1,p2,p3]) 
```

```python
doc1 = nlp(u"The Solar Power industry continues to grow as demand for solarpower increase.Solar-Power cars are gaining popularity.")
```

```python
matches = matcher(doc1)
print(matches)
```

```python
for matchid,start,end in matches:
    stringid = nlp.vocab.strings[matchid]
    span = doc1[start:end]
    print(matchid,stringid,start,end,span.text)
```

```python
p4 = [{'LOWER':'solarpowered'}]
matcher.add('SolarPower',[p1,p2,p3,p4])
doc2 = nlp(u"Solar-Powered energy runs solar-powered cars")
matches2 = matcher(doc2)
print(matches2)
```

```python
from spacy.matcher import PhraseMatcher
matcher = PhraseMatcher(nlp.vocab)

with open(r'reaganomics.txt',encoding='cp1252') as f:
    doc3=nlp(f.read())
    
    
```

```python
phraselist = ['voodoo economics','supply-side economics','trickle-down economics','free-market economics']

phrasepatterns = [nlp(text) for text in phraselist]

matcher.add('VoodooEconomics',None, *phrasepatterns)
matches = matcher(doc3)
```

```python
for i in matches:
    print(i)
```
