```python
txt = "The agent's phone number is 34344-45454 . call soon"
```

```python
"phone" in txt
```

```python
import re
pattern = "phone"
match = re.search(pattern,txt)
```

```python
print(match)
```

```python
txt = "My phone is a new phone"
match = re.search(pattern,txt)
print(match)
```

```python
matches = re.findall("phone",txt)
print(matches)
```

```python
for match in re.finditer("phone",txt):
    print(match.span())
```
