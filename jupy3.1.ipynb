```python
import spacy
spacy.cli.download("en_core_web_md")
```

```python
import spacy
nlp = spacy.load("en_core_web_md")
```

```python
print(nlp(u"lion").vector)
```

```python
doc = nlp(u"The quick brown fox jumped over the lazy dogs")
doc.vector
```

```python
tokens = doc
for t1 in tokens:
    for t2 in tokens:
        print(t1.text,t2.text,t1.similarity(t2))
```
